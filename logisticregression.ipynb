```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mushroom_ds = pd.read_csv("/kaggle/input/mushroom-dataset/mushroom_cleaned.csv", names=['capDiameter', 'capShape', 'gillAttachment', 'gillColor', 'stemHeight', 'stemWidth', 'stemColor', 'season', 'class'], header=None)
mushroom_ds = mushroom_ds[1::]
mushroom_ds.head()
```

```python
X = mushroom_ds[['capDiameter', 'capShape', 'gillAttachment', 'gillColor', 'stemHeight', 'stemWidth', 'stemColor', 'season']]
y = mushroom_ds['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=33)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train[0:5])
```

```python
logReg = LogisticRegression(solver="sag")
logReg.fit(X_train, y_train)
y_pred = logReg.predict(X_test)
```

```python
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
```

```python
accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```
